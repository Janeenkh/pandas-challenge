```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_type_df = school_data_complete.groupby(["school_name"])["type"].unique().to_frame()
grouped = school_data_complete.groupby(["school_name"])
school_type_df
```

```python
students_df = grouped ["Student ID"].count().to_frame()
students_df
```

```python
budget_df = grouped ["budget"].unique().to_frame()
budget_df
```

```python
#perbudget_df = (budget_df['budget']/students_df['Student ID']).to_frame()
#perbudget_df = perbudget_df.rename({"0":"per student budget"},axis=1)
#perbudget_df
```

```python
per_school_counts = school_data_complete["school_name"].value_counts()
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
perbudget_df = per_school_budget / per_school_counts
perbudget_df = pd.Series(perbudget_df, name="per student budget").to_frame()
#perbudget_df= pd.DataFrame(perbudget_df)
#perbudget_df.rename(columns={"per student budget"}, inplace=True)
#perbudget_df.to_frame()
perbudget_df = perbudget_df.rename_axis('school_name')
perbudget_df = pd.DataFrame(perbudget_df)
#perbudget_df = perbudget_df.rename({"0":"hi"},axis=1)

perbudget_df
```

```python
math_score_df = grouped ["math_score"].mean().to_frame()
math_score_df = math_score_df.rename({"math_score":"avg math score"},axis=1)
math_score_df
```

```python
reading_score_df = grouped ["reading_score"].mean().to_frame()
reading_score_df = reading_score_df.rename({"reading_score":"avg reading score"},axis=1)
reading_score_df
```

```python
passmathcount_df = school_data_complete.loc[school_data_complete["math_score"]>=70, "Student ID"].count()
math_pass_df = passmathcount_df/students_df
math_pass_df = pd.DataFrame(math_pass_df)
math_pass_df = math_pass_df.rename({"Student ID":"% passing math"},axis=1)
math_pass_df
```

```python
passreadcount_df = school_data_complete.loc[school_data_complete["reading_score"]>=70, "Student ID"].count()
reading_pass_df = passreadcount_df/students_df
reading_pass_df = pd.DataFrame(reading_pass_df)
reading_pass_df = reading_pass_df.rename({"Student ID":"% pass reading"},axis=1)
reading_pass_df
```

```python
overall_df = school_data_complete.loc[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70), "Student ID"].count()
overall_pass_per_df = overall_df/students_df
overall_pass_per_df = pd.DataFrame(overall_pass_per_df)
overall_pass_per_df = overall_pass_per_df.rename({"Student ID":"% overall passing"},axis=1)
overall_pass_per_df
```

```python
merge_df = pd.merge(reading_pass_df, math_score_df, on="school_name")
merge_df = pd.merge(merge_df, reading_score_df, on="school_name")
merge_df = pd.merge(merge_df, math_pass_df, on="school_name")

merge_df = pd.merge(merge_df, perbudget_df, on="school_name")
merge_df = pd.merge(merge_df, budget_df, on="school_name")
merge_df = pd.merge(merge_df, students_df, on="school_name")

merge_df = pd.merge(merge_df, overall_pass_per_df, on="school_name")

merge_df
```

## Math Scores by Grade

```python
nineth=school_data_complete.loc[school_data_complete['grade'] == '9th'].groupby('school_name')['math_score'].mean().to_frame().reset_index()
nineth=nineth.rename(columns={'math_score': '9th'})
```

```python
tenth=school_data_complete.loc[school_data_complete['grade'] == '10th'].groupby('school_name')['math_score'].mean().to_frame().reset_index()
tenth=tenth.rename(columns={'math_score': '10th'})
```

```python
eleventh=school_data_complete.loc[school_data_complete['grade'] == '11th'].groupby('school_name')['math_score'].mean().to_frame().reset_index()
eleventh=eleventh.rename(columns={'math_score': '11th'})
```

```python
twelveth=school_data_complete.loc[school_data_complete['grade'] == '12th'].groupby('school_name')['math_score'].mean().to_frame().reset_index()
twelveth=twelveth.rename(columns={'math_score': '12th'})
```

```python
math_scores_by_grade=nineth.merge(tenth, on="school_name").merge(eleventh,on ="school_name").merge(twelveth, on="school_name")
math_scores_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins=[0, 585, 630, 645, 680]
labels=['>585','585-630', '630-645','645-680' ]
```

```python
merge_df['Spending'] = pd.cut(merge_df['per student budget'], bins=bins, labels = labels )

merge_df

scores_by_school_spending=merge_df.groupby('Spending').agg({'% passing math' : 'mean', 'avg reading score' : 'mean', 'avg math score' : 'mean', '% overall passing' : 'mean'})

scores_by_school_spending




```
